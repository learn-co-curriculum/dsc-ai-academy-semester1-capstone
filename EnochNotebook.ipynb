## Top 10 Grossing movies in each year

```python
fig, axs = plt.subplots()

axs.bar(movie_gross2010['title'],
         movie_gross2010['domestic_gross'],
         color='g', align='center')
axs.set_title('Top 10 Grossing Movies 2010')
axs.set_xlabel('Movie')
axs.set_ylabel('Gross(millons)')
axs.set_xticklabels(movie_gross2010['title'], rotation=45)
axs.tick_params(axis="both", direction="out", pad=15)




```

```python
fig, axs = plt.subplots(1, 3, figsize=(30, 12), sharey=True)
axs[1].bar(movie_gross2010['title'],
         movie_gross2010['domestic_gross'],
         color='g', linestyle='dashdot')
axs[1].set_title('Shampoo Sales Across 3 Years')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Total Sales')


axs[2].bar(movie_gross2010['title'],
         movie_gross2010['domestic_gross'],
         color='g', linestyle='dashdot')
axs[2].set_title('Shampoo Sales Across 3 Years')
axs[2].set_xlabel('Year')
axs[2].set_ylabel('Total Sales')


fig.suptitle('Categorical Plotting')
```
